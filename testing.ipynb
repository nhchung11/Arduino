```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 100
N = [80, 85, 90, 95, 100, 105, 110, 115, 120]

def epsilon(M, N):
    result = []
    for i in range(len(N)):
        tmp = (M / N[i]) - 1
        result.append(tmp)
    return result

Gm = []
for i in range(len(N)):
    tmp = M / N[i]
    Gm.append(tmp)
print(epsilon(M, N))
```

```python
G = []
for i in range(len(N)):
    tmp = np.sum(Gm[i]) / M
    G.append(tmp)

Pr = []
for i in range(len(N)):
    tmp = 1 - np.exp(-1 * (1 + epsilon(M, N)[i]) * G[i])
    Pr.append(tmp)

plt.plot(N, Pr)
plt.show()
```
